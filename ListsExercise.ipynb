```python
# 3 built in methods for Lists:

list = ["apple", "banana", "cherry", "banana"]

# count the occurences of a given element in the list
print(list.count("banana"))


# find the first index of a given element in the list
print(list.index("banana")) 


# reverse the elements of the list
list.reverse()
print(list)
```

```python
# 3 built in methods for Dictionaries:

student = {"name": "Panha", "age": 21, "city": "Phnom Penh"}

# returns the value for the specified key
student.get("name")  
print(student.get("name"))

# remove and return the most recently added key-value pair in the dict
print(student.popitem())

# removes all items from the dictionary
student.clear()
print(student)

```

```python
# create a list with only temps above 30 degrees
temperatures = [28, 32, 25, 30, 35, 29, 31]
filtered = []
for i in temperatures:
    if i > 30:
        filtered.append(i)

print(filtered)

filtered = [i for i in temperatures if i > 30]
print(filtered)
```

```python
# create a dictionary to store occurences of an element in a list

colors = ["red", "blue", "red", "green", "blue", "red", "yellow"]
dict = {}

# dict["red"] = colors.count("red")
# dict["blue"] = colors.count("blue")
# dict["green"] = colors.count("green")
# dict["yellow"] = colors.count("yellow")

for color in colors:
    dict[color] = dict.get(color, 0) + 1
    
print(dict)

dict = {}
```

```python
users = [
    { "name": "Alex", "age": 28, "status": "active" },
    { "name": "Ben", "age": 19, "status": "inactive" },
    { "name": "Chloe", "age": 35, "status": "active" },
    { "name": "Dana", "age": 22, "status": "inactive" },
    { "name": "Ethan", "age": 40, "status": "active" },
]

totalage = 0

for i in users:
    for key, value in i.items():
        if i["status"] == "active":
            if key == "age":
                totalage += value

print(totalage)
```

```python
totalage = sum(user["age"] for user in users if user["status"] == "active")
print(totalage)
```

```python
#

inventory = [
    { "id": 101, "name": "Laptop Stand", "price": 45.00 },
    { "id": 202, "name": "Wireless Mouse", "price": 25.50 },
    { "id": 303, "name": "Mechanical Keyboard", "price": 120.00 }, 
]
shoppingCart = [202, 101, 202] #IDs in the cart
new_cart = []

for i in shoppingCart:
    for object in inventory:
        if object["id"] == i:
            new_cart.append(object["price"])

print(new_cart)

def f(cart_id):
    for item in inventory:
        if item["id"] == cart_id:
            return item["price"]
        

map_cart = list(map(f, shoppingCart))
print(map_cart)


```
